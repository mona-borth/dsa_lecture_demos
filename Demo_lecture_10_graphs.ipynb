# Breadth-first Search Algorithm
Breadth-First Search (BFS) is an algorithm for traversing or searching graph data structures. It starts at a selected node (called the "root") and explores all its neighboring nodes at the current depth before moving on to nodes at the next depth level.

It is implemented as follows:
1. start from node s
1. Initialize an empty queue and add the starting node s
1. Initialize an empty visited list
1. While the queue is not empty:
  - Dequeue the front node u
  - If u is not visited:
    - Mark u as visited
    - Enqueue all unvisited neighbors of u

Let us implement the above sudo-code in python:

````python
import matplotlib.pyplot as plt
import networkx as nx

def bfs(graph, start):
    """
    Perform BFS on a graph from a starting node using a simple list.

    Parameters:
    - graph: Adjacency list representation of the graph (dict)
    - start: Starting node for BFS

    Returns:
    - visited: List of nodes in the order they were visited
    """
    visited = []
    queue = [start]

    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.append(node)
            for neighbor in graph[node]:
                if neighbor not in visited and neighbor not in queue:
                    queue.append(neighbor)

    return visited

# Example usage
graph_example = {
    0: [1, 2, 5],
    1: [0, 3, 4],
    2: [0],
    3: [1],
    4: [1],
    5: [0]
}

start = 1
print("BFS traversal:", bfs(graph_example, start))

# Visualization
G = nx.Graph()

# Add edges
for node in graph_example:
    for neighbor in graph_example[node]:
        G.add_edge(node, neighbor)

# Draw graph
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=800, font_size=14, font_weight='bold', edge_color='gray')
plt.title("Graph Visualization with Labeled Nodes")
plt.show()
````

What is your analysis of time complexity?
<details>
  <summary>Answer</summary>
</p>$O(n + m)$: Where n and m are the lengths of the two arrays. Each element (node or edge) is visited once.</p>
</details>

# Depth-First Search (DFS) Algorithm
Depth-First Search (DFS) is an algorithm for traversing or searching graph data structures. It starts at the root node and explores as far as possible along each branch before backtracking.

It is implemented as follows:
1. Start from node s
1. Initialize an empty stack and push the starting node s onto it
1. Initialize an empty visited list
1. While the stack is not empty:
  - Pop the top node u from the stack
  - If u is not in visited:
    - Mark u as visited
    - Push all unvisited neighbors of u onto the stack

Let us take a look on the python implementation:

````python
import matplotlib.pyplot as plt
import networkx as nx

def dfs_non_recursive(graph, start):
    """
    Perform non-recursive DFS on a graph from a starting node.

    Parameters:
    - graph: Adjacency list representation of the graph (dict)
    - start: Starting node for DFS

    Returns:
    - visited: List of nodes in the order they were visited
    """
    visited = []
    stack = [start]

    while stack:
        node = stack.pop()
        if node not in visited:
            visited.append(node)
            # Add neighbors to the stack in reverse order for consistent traversal
            stack.extend(reversed([n for n in graph[node] if n not in visited]))

    return visited

# Example usage
graph_example = {
    0: [1, 2, 5],
    1: [0, 3, 4],
    2: [0],
    3: [1],
    4: [1],
    5: [0]
}
start = 1
print("Non-Recursive DFS traversal:", dfs_non_recursive(graph_example, start))


# Visualization
G = nx.Graph()

# Add edges
for node in graph_example:
    for neighbor in graph_example[node]:
        G.add_edge(node, neighbor)

# Draw graph
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=800, font_size=14, font_weight='bold', edge_color='gray')
plt.title("Graph Visualization with Labeled Nodes")
plt.show()
````

Alternatively one can implement DFS in a recursive manner using:

It is implemented as follows:
1. Start from node s
1. Initialize an empty visited list
1. Define a recursive function:
  - If node u is not visited:
    - Mark u as visited
    - Recursively call the function for each unvisited neighbor of u
1. Call the recursive function with starting node s

Let us see how it looks like in python.

````python
import matplotlib.pyplot as plt
import networkx as nx

def dfs(graph, start, visited=None):
    """
    Perform DFS on a graph from a starting node.

    Parameters:
    - graph: Adjacency list representation of the graph (dict)
    - start: Starting node for DFS
    - visited: Set of visited nodes

    Returns:
    - visited: List of nodes in the order they were visited
    """
    if visited is None:
        visited = []

    if start not in visited:
        visited.append(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

    return visited

# Example usage
graph_example = {
    0: [1, 2, 5],
    1: [0, 3, 4],
    2: [0],
    3: [1],
    4: [1],
    5: [0]
}

start = 3
print("DFS traversal:", dfs(graph_example, start))


# Visualization
G = nx.Graph()

# Add edges
for node in graph_example:
    for neighbor in graph_example[node]:
        G.add_edge(node, neighbor)

# Draw graph
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=800, font_size=14, font_weight='bold', edge_color='gray')
plt.title("Graph Visualization with Labeled Nodes")
plt.show()
````

What is your analysis of time complexity?
